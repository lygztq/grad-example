As shown in [Pytorch's Doc](https://docs.pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html), the formular of layernorm is $$y=\dfrac{x-\text{E}[x]}{\sqrt{\text{Var}(x)+\epsilon}} * \gamma + \beta.$$

However, this formular is only vaild for scalar scenario. For matrix input, this is invalid. Here we give the matrix formular.

Support the weights of layernorm are gamma and beta, all d-dim row vector $\R^{1 \times d}$. And suppose the normalized input is $z\in \R^{N\times d}$, the layernorm can be written as $$y=z \odot \mathbf{1}_{N\times 1}\gamma + \mathbf{1}_{N\times 1}\beta.$$

The $\mathbf{1}_{N\times 1}\gamma$ actually broadcast gamma along the first dim. Now according the for total differetial formular, given a loss $\mathcal{L}$, we have $$\text{d}\mathcal{L} = \text{tr}(\dfrac{\partial \mathcal{L}}{\partial y}^{\top}\text{d}y),$$ and for the two weight of layernorm, we have $$\text{d}y = \text{d}z \odot \mathbf{1}_{N\times 1}\gamma + z \odot \mathbf{1}_{N\times 1}\text{d}\gamma + \mathbf{1}_{N\times 1}\text{d}\beta.$$

For the beta part, we have $$\text{d}\mathcal{L} = \text{tr}(\dfrac{\partial \mathcal{L}}{\partial y}^{\top}\mathbf{1}_{N\times 1}\text{d}\beta),$$ thus $$\dfrac{\partial \mathcal{L}}{\partial \beta} = \mathbf{1}_{N\times 1}^{\top} \dfrac{\partial \mathcal{L}}{\partial y}.$$ The $\mathbf{1}_{N\times 1}^{\top} \dfrac{\partial \mathcal{L}}{\partial y}$ is actually reduce $\dfrac{\partial \mathcal{L}}{\partial y}$ along the first dim (N-dim), which means we sum all gradients from all output tokens to get the gradient of beta.
